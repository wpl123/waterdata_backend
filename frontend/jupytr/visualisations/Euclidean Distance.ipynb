```python
d1 = "If you fail to plan then you plan to fail"
d2 = "do not fail me or else I will fail you"
d3 = "I always organise well"


def cosine_similarity(vector1, vector2):
    assert (len(vector1) == len(vector2)),"The two vectors do not have the same length"
    #cosine = 0
    cosine = np.inner(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
    return cosine

                                                                                
# Insert some code here to calculate the vocabulary
# Calculate the vocabulary
d1_words = d1.split(" ")
d1_voc = set()
for d1_word in d1_words:
    d1_voc.add(d1_word)
    print(d1_voc)
#
#Calculate the vocabulary
d2_words = d2.split(" ")
d2_voc = set()

for d2_word in d2_words:
    d2_voc.add(d2_word)
print(d2_voc)

                                                                        
# Calculate the vocabulary
d3_words = d3.split(" ")
d3_voc = set()
for d3_word in d3_words:
    d3_voc.add(d3_word)
                                                                                
print(d3_voc)

# Vocabulary
V = d1_voc.union(d2_voc)
V = V.union(d3_voc)
print(V)

# Insert some code here to represent each document against the vocabulary of the collection
d1_bin = [];
for item in V:
    if item in d1_voc:
        d1_bin.append(1)
    else:
        d1_bin.append(0)
                                                                                
print(d1_bin)

d2_bin = [];
for item in V:
    if item in d2_voc:
        d2_bin.append(1)
    else:
        d2_bin.append(0)
                                                                                
print(d2_bin)
d3_bin = [];
for item in V:
    if item in d3_voc:
        d3_bin.append(1)
    else:
        d3_bin.append(0)
                                                                                
print(d3_bin)
               
import numpy as np

# Calculate the Euclidean similarity between d1 and d2, and between d1 and d3
Euclidean_dist_d1d2 = np.linalg.norm(np.array(d1_bin)-np.array(d2_bin))
print('Euclidean distance of d1 and d2',Euclidean_dist_d1d2)
Euclidean_dist_d1d3 = np.linalg.norm(np.array(d1_bin)-np.array(d3_bin))
print('Euclidean distance of d1 and d3',Euclidean_dist_d1d3)
                                                                                
# Calculate the cosine similarity between d1 and d2, and between d1 and d3
cosine_sim_d1d2 = cosine_similarity(d1_bin, d2_bin)
print('Cosine similarity of d1 and d2',cosine_sim_d1d2)
cosine_sim_d1d3 = cosine_similarity(d1_bin, d3_bin)
print('Cosine similarity of d1 and d3',cosine_sim_d1d3)
```
